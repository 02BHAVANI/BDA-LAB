```python
# Data visualization
plt.figure(figsize=(10, 6))
# Replace 'Age' with the actual numerical column you want to plot
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Histogram of Age')
plt.show()
```

```python
# Data visualization
plt.figure(figsize=(10, 6))
# Replace 'your_actual_column_name' with the actual column name from your DataFrame
sns.countplot(x='Age', data=df)
plt.title('Count Plot of Categorical Column')
plt.xticks(rotation=45)
plt.show()
```
